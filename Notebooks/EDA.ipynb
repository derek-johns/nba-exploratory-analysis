# Exploratory Data Analysis for NBA 2020-21 Season  
* **Data Source:** [nba-api](https://github.com/swar/nba_api)

## 1. Classifying Statistics  

* We will first attempt to understand which stats are most relevant in evaluating player and team metrics.  


* Most NBA stats will fall into either one of two categories, **Production** or **Context**.  


* These categories are each composed of sub-categories that will help us identify the role each stat plays in determining player and team success.  

| **Production** | **Context** |
| :------------: | :---------: |
| Quantity       |  Usage      | 
| Efficiency     |  Teammates  |
| Type           |  Opponents  |
| Versatility    |  Situation  |

### Quantity
* The "how much" of production.  


* Include stats such as PTS, AST, REB, etc.   


* Overemphasized because these stats lack context. We want to know more than just how many points a player scored. 
    * How did they score those points? (Type) 
    * How many shots did they take to score those many points? (Efficiency) 
    * Were these shots closely contested? (Situation)
    * How good were the other players on the court when they scored these points? (Teammates/Opponents)

### Efficiency
* Production per opportunity.


* Include stats such as FG%, FG3%, FT%, etc.


* **Effective Field Goal Percentage (EFG%)** is similar to FG% but accounts for the added value of 3 Point Field Goals Made (FG3M).



* **True Shooting Percentage (TS%)** goes a step further then EFG% and also accounts for Free Throws (FTA).



# Glossary

**AST** : Assists

**FG%** : Field Goal Percentage

**FGA** : Field Goals Attempted

**FGM** : Field Goals Made

**FG3%** : 3 Point Field Goal Percentage

**FG3A** : 3 Point Field Goals Attempted

**FG3M** : 3 Point Field Goals Made

**FT%** : Free Throw Percentage

**PTS** : Points  

**REB** : Rebounds  

**TS%** : True Shooting Percentage
